### .apply( )

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"

bond["Box_Office"] = bond["Box_Office"].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis="columns") #move along columns -> the function operates along columns for each row
```
