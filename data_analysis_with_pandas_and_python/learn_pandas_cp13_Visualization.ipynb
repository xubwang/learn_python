### bar graph

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_performance).value_counts()
```

```python
#plt.style.use("ggplot")
cat_data = bb["Close"].apply(rank_performance).value_counts()
cat_data.plot(kind="bar")
```

```python
cat_data.plot(kind="barh")
```
