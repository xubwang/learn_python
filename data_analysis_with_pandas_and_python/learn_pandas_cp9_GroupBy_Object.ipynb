### .agg()

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Industry")
fortune.head(3)
```

```python
sectors.agg({"Revenue": "sum",
            "Profits": "mean"})
```

```python
sectors.agg(["mean", "sum"])
```

```python
sectors[["Revenue", "Profits"]].agg(["mean", "sum", "max", "size"])
```

```python
sectors.agg({"Revenue": ["sum", "max"],
            "Profits": "mean"})
```

### iterating through groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors["Profits"].max()
```

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
# i = 0
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = pd.concat([df, highest_revenue_company_in_group])
#     df.loc[i] = highest_revenue_company_in_group
#     i+=1
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = pd.concat([df, highest_revenue_in_city])
```
